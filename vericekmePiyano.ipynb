```python
import yt_dlp
import os

# İndirme ve işleme yapılacak klasör
output_folder = "piyano_sarkilari"

# Klasörü oluştur (varsa hata vermez)
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# YouTube playlist  tüm videoları indir ve ses dosyasını çıkart
def download_playlist_audio(playlist_url, output_folder):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': os.path.join(output_folder, '%(title)s.%(ext)s'),
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'noplaylist': False,  # Çalma listesi indirilmesini sağlamak için False olmalı
        'ignoreerrors': True,  # Hataları yok say ve devam et
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([playlist_url])
        print(f"Çalma listesi başarıyla indirildi ve {output_folder} klasörüne kaydedildi.")
    
    except Exception as e:
        print(f"Bir hata oluştu: {str(e)}")

#  YouTube playlist  URL'si
playlist_url = "https://www.youtube.com/playlist?list=PLEJaoODCJ8rOJ4p6wzPCZXUC5UtqiULek"  

# Playlist'teki tüm videoları indir ve MP3 formatında kaydetme
download_playlist_audio(playlist_url, output_folder)
```
